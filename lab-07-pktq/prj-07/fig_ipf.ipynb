# General

```python
# OK
# import packages

import os 
import regex as re
import numpy as np 
from matplotlib import pyplot as plt, gridspec as gridspec # local: 3.4.1
```

```python
# OK: EXTRACT FUNCTIONS

def extract_txt_for_bw(data_dir):
    path = "./" + data_dir + "/" + "iperf_result.txt"
    
    pat_va = pat_bandwidth

    l_ts = []
    l_va = []
    count = 0.05
    with open(path, 'r') as file:
        logs = file.readlines()
        for log in logs:
            # pattern match
            if not re.match(ipf_head, log):
                continue

            va_match_list = re.findall(pat_va, log)
            if len(va_match_list) < 1:
                continue
            
            # store va value
            va= float(va_match_list[0])
            
            l_ts.append(count)
            l_va.append(va)
            
            count += 0.1

    # test or true return
    return l_ts, l_va


# FOR MT TRANSFER SIZE ONLY

pat_transfer_K = "([0-9]*[\.0-9]*)\sKBytes"
pat_transfer_M = "([0-9]*[\.0-9]*)\sMBytes"

def extract_txt_for_trans(data_dir):
    path = "./" + data_dir + "/" + "iperf_result.txt"

    l_ts = []
    l_va = []
    count = 0.05
    with open(path, 'r') as file:
        logs = file.readlines()
        for log in logs:
            # pattern match
            if not re.match(ipf_head, log):
                continue

            va_match_list_k = re.findall(pat_transfer_K, log)
            va_match_list_m = re.findall(pat_transfer_M, log)
            if len(va_match_list_k) < 1:
                if len(va_match_list_m) < 1:
                    continue
                else:
                    va= float(va_match_list_m[0])*1000
            else:
                va= float(va_match_list_k[0])

            l_ts.append(count)
            l_va.append(va)
            
            count += 0.1

    # test or true return
    return l_ts, l_va
```

```python
ipf_head = "\[[\s0-9]+\]"
va_type = ["Transfer", "Bandwidth"]

pat_bandwidth = "([0-9]*[\.0-9]*)\sMbits/sec"

folder_qlen="qlen-.+[0-9]+"
folder_algo="algo-[a-z]+"
```

## ALGO

```python
# list folders, select the ones that store the data output
path = '.'
items = os.listdir(path)
data_dirs = []
for item in items:
    # item is directory && 'xxxx' in item
    if os.path.isdir(item) and re.match(folder_algo, item):
        data_dirs.append(item)
```

```python
# BANDWIDTH
type = va_type[1]
color = 'lightseagreen'
    
fig, ax = plt.subplots(1, 3) 
count=0
for res in data_dirs:
    x, y = extract_txt_for_bw(res)
    ax[count].scatter(x, y,
                label=type,
                c=color,
                s=10,
                alpha=0.5, 
                edgecolors='none')
    ax[count].title.set_text('algo='+res[5:])
    ax[count].set_xlim([0,60])
    ax[count].grid(True)
    count+=1

# so far so good, now the trick
# lines_labels = [a.get_legend_handles_labels() for a in ax]
# lines, labels = [sum(lol, []) for lol in zip(*lines_labels)]
lines, labels = ax[0].get_legend_handles_labels()

# finally we invoke the legend (that you probably would like to customize...)
fig.legend(
    lines, labels, 
    bbox_to_anchor = (0,0.1,1,1),
    loc='upper right')

plt.tight_layout()
plt.subplots_adjust(wspace=0.35, hspace=0.001)

fig_path = "./figs_mt/iperf-%s-fig.png" % (type)
plt.savefig(fig_path, dpi=800, 
            bbox_inches = 'tight') # ensure legend output correctly
plt.close()
```

```python
# TRANSFER
type = va_type[0]
color = 'darkorange'

fig, ax = plt.subplots(1, 3) 
count=0
for res in data_dirs:
    x, y = extract_txt_for_trans(res)
    ax[count].scatter(x, y,
                label=type,
                c=color,
                s=10,
                alpha=0.5, 
                edgecolors='none')
    ax[count].title.set_text('algo='+res[5:])
    ax[count].set_xlim([0,60])
    ax[count].grid(True)
    count+=1

# so far so good, now the trick
# lines_labels = [a.get_legend_handles_labels() for a in ax]
# lines, labels = [sum(lol, []) for lol in zip(*lines_labels)]
lines, labels = ax[0].get_legend_handles_labels()

# finally we invoke the legend (that you probably would like to customize...)
fig.legend(
    lines, labels, 
    bbox_to_anchor = (0,0.1,1,1),
    loc='upper right')

plt.tight_layout()
plt.subplots_adjust(wspace=0.35, hspace=0.001)

fig_path = "./figs_mt/iperf-%s-fig.png" % (type)
plt.savefig(fig_path, dpi=800, 
            bbox_inches = 'tight') # ensure legend output correctly
plt.close()
```
